```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from src.style_time import style_time
```

# Google Stocks

## Normal 

```python
in_sample_train = np.load("data/google_stocks/preprocessed/in_sample_train.npy")
in_sample_test = np.load("data/google_stocks/preprocessed/in_sample_test.npy")

transfered_in_samples_train = np.load("data/google_stocks/preprocessed/in_sample_train_transfered.npy")
transfered_in_samples_test = np.load("data/google_stocks/preprocessed/in_sample_test_transfered.npy")

style_train = np.load("data/google_stocks/preprocessed/style_train.npy")
style_test = np.load('data/google_stocks/preprocessed/style_test.npy')

ALPHA = 10
BETA = 50
GAMMA = 1
LR = 0.01
EPOCHS = 500
```

```python
def plot_result(content_sequence, style_sequence, gen_sequence):

    plt.figure(figsize=(18, 10))
    ax = plt.subplot(311)
    ax.set_title("Content Sequence")

    ax.plot(content_sequence)
    ax.grid()

    ax = plt.subplot(312)
    ax.set_title("Style Sequence")

    ax.plot(style_sequence)
    ax.grid()

    ax = plt.subplot(313)
    ax.set_title("Generated Sequence")

    ax.plot(gen_sequence)
    ax.grid()


idx = 1

content_sequence = in_sample_train[idx]
style_sequence = style_train[idx]

generated_sequences, c_losses, s_losses, total_losses = style_time(content_sequence, style_sequence, EPOCHS, verbose=1, alpha=ALPHA, beta=BETA, gamma=GAMMA, learning_rate=LR)

plot_result(c_losses, s_losses, total_losses)

plot_result(content_sequence[0, :], style_sequence[0, :], generated_sequences[0, :])
```
